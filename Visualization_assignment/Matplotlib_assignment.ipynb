#### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.figure(figsize=(8,4))
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Scatter plot")
plt.scatter(x,y);
```

#### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

The subplot() function in Python is used to create a grid of subplots within a single figure. It allows us to arrange multiple plots in a grid-like structure, making it easier to compare and visualize different data or relationships.

```python
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
fig,ax=plt.subplots(2,2)
ax[0,0].plot(x1,y1)
ax[0,1].plot(x2,y2)
ax[1,0].plot(x3,y3)
ax[1,1].plot(x4,y4)
plt.tight_layout()
plt.show()
```

#### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

A bar plot, also known as a bar chart or bar graph, is a type of visualization that represents categorical data with rectangular bars. The length of each bar corresponds to the quantity or frequency of the data it represents. Bar plots are commonly used to compare different categories or to display the distribution of a categorical variable.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.figure(figsize=(6,4))
plt.bar(company,profit);
```

```python
plt.figure(figsize=(6,4))
plt.barh(company,profit);
```

#### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

A box plot, also known as a box-and-whisker plot, is a type of statistical visualization that provides a summary of the distribution of a dataset. It displays the minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum values of a dataset.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.boxplot(box1);
```

```python
plt.boxplot(box2);
```
