#### Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a python based framework which is used to create APIs or build web applications. It has its own libraries and functions to develop web applications.

Advantages:
 1. Flask has a very simple UI therefore it is very easy to develop apps using flask.
 
 2. Flask comes with a built-in development server, allowing you to quickly test and debug your applications without the need for additional setup. 

 3. Flask uses the Jinja2 templating engine, which provides powerful and flexible templating capabilities. This makes it easy to separate the presentation logic from the application logic, resulting in cleaner and more maintainable code.

#### Q3. What is App routing in Flask? Why do we use app routes?

App routing is a process of associating a URL endpoint with a specific function or view in our Flask application.It allows you to define different routes for different parts of your application and determine what function should be executed when a specific URL is accessed.

#### Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/” route to show the following details:
#### Company Name: ABC Corporation
#### Location: India
#### Contact Detail: 999-999-9999

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home_page():
    return '<h1>Company Name: ABC Corporation <br>Location: India <br>Contact Detail: 999-999-9999</h1' 

@app.route('/welcome')
def greeting():
    return '<h2>Welcome to ABC Corporation</h2>'
```

#### Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

To perform URL building in Flask, you can use the url_for() function. This function generates a URL for the specified endpoint.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home_page():
    return '<h1>Company Name: ABC Corporation <br>Location: India <br>Contact Detail: 999-999-9999</h1' 

@app.route('/welcome')
def greeting():
    return '<h2>Welcome to ABC Corporation</h2>'

with app.test_request_context():
    # Generate URL for the 'greeting' endpoint
    url = url_for('greeting')
    print(f'URL for greeting: {url}')
```
